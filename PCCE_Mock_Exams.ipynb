```python
# [ë¬¸ì œ 4] ë¬¸ìì—´ ì••ì¶• (Logic)
# 'aaabbb' -> 'a3b3' ë¡œ ë³€í™˜í•˜ëŠ” ê°„ë‹¨í•œ ë¡œì§
s = "aaabbb"
res = ""
cnt = 1
for i in range(1, len(s)):
    if s[i] == s[i-1]:
        cnt += 1
    else:
        res += s[i-1] + str(cnt)
        cnt = 1
res += s[-1] + str(cnt)
print(res)
        
```

```python
# [ë¬¸ì œ 5] 2ì°¨ì› ë¦¬ìŠ¤íŠ¸ í•©ê³„
# 3x3 ë¦¬ìŠ¤íŠ¸ì˜ ëª¨ë“  ìš”ì†Œì˜ í•©ì„ êµ¬í•˜ì„¸ìš”.
matrix = [[1,2,3], [4,5,6], [7,8,9]]
total = 0
for row in matrix:
    total += sum(row)
print(total)
```

```python
# [ë¬¸ì œ 8] [ë¹ˆì¹¸] ëŒë‹¤ ì •ë ¬
# ë¬¸ìì—´ ê¸¸ì´ ìˆœìœ¼ë¡œ ì •ë ¬í•˜ì„¸ìš”.
words = ['apple', 'bat', 'car', 'banana']

# words.sort(key=...) : ì •ë ¬ ê¸°ì¤€ì„ ì§ì ‘ ì„¤ì •í•©ë‹ˆë‹¤.
# lambda x: len(x) : ë¦¬ìŠ¤íŠ¸ì˜ ê° ìš”ì†Œ(x)ë¥¼ 'ê¸¸ì´(len)'ë¡œ ë°”ê¿”ì„œ ë¹„êµí•˜ê² ë‹¤ëŠ” ëœ»!
# ê²°ê³¼: ê¸¸ì´ê°€ ì§§ì€ ìˆœì„œëŒ€ë¡œ ì •ë ¬ë¨ (bat(3) -> car(3) -> apple(5) -> banana(6))
words.sort(key=lambda x: len(x))
print(words)
```

```python
# [ë¬¸ì œ 9] [ì‹œë®¬ë ˆì´ì…˜] ë¡œë´‡ ì²­ì†Œê¸° (ë°©í–¥ íšŒì „)
# í˜„ì¬ ë°©í–¥ì—ì„œ ì™¼ìª½ìœ¼ë¡œ íšŒì „(90ë„ ë°˜ì‹œê³„) í›„ ì „ì§„
# 0:ë¶, 1:ë™, 2:ë‚¨, 3:ì„œ
d = 0 # ë¶ìª½

# [íšŒì „ ê³µì‹ ì„¤ëª…]
# d - 1 : í˜„ì¬ ë°©í–¥(0)ì—ì„œ 1ì„ ëºë‹ˆë‹¤ -> -1 (ë°˜ì‹œê³„ ë°©í–¥)
# (d - 1) % 4 : -1ì„ 4ë¡œ ë‚˜ëˆˆ ë‚˜ë¨¸ì§€ -> 3 (ì„œìª½)
# íŒŒì´ì¬ì˜ % ì—°ì‚°ìëŠ” í•­ìƒ ì–‘ìˆ˜ ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ë¯€ë¡œ -1 % 4 = 3ì´ ë©ë‹ˆë‹¤.
d = (d - 1) % 4
print(f"ì™¼ìª½ íšŒì „ í›„ ë°©í–¥: {d} (3:ì„œìª½)")
```

```python
# [ë¬¸ì œ 10] [ì•Œê³ ë¦¬ì¦˜] ì†Œìˆ˜ íŒë³„ (Prime Check)
# ì •ìˆ˜ Nì´ ì†Œìˆ˜ì¸ì§€ íŒë³„í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‘ì„±í•˜ì„¸ìš”.
def is_prime(n):
    if n < 2: return False # 2ë³´ë‹¤ ì‘ìœ¼ë©´ ì†Œìˆ˜ê°€ ì•„ë‹˜ (0, 1 ì œì™¸)
    
    # 2ë¶€í„° ë£¨íŠ¸ nê¹Œì§€ì˜ ìˆ˜ë¡œ ë‚˜ëˆ ë´…ë‹ˆë‹¤.
    # ì™œ ë£¨íŠ¸ nê¹Œì§€ë§Œ í•˜ë‚˜ìš”? -> ê·¸ë³´ë‹¤ í° ì•½ìˆ˜ëŠ” ì–´ì°¨í”¼ ì‘ì€ ì•½ìˆ˜ì™€ ì§ê¿ì´ë¼ ì´ë¯¸ ê±¸ëŸ¬ì§€ê¸° ë•Œë¬¸!
    # 36ì˜ ì•½ìˆ˜: (1, 36), (2, 18), (3, 12), (4, 9), (6, 6) <- 6ê¹Œì§€ë§Œ ë³´ë©´ ë¨
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: # ë‚˜ëˆ„ì–´ ë–¨ì–´ì§€ë©´(ì•½ìˆ˜ê°€ ìˆìœ¼ë©´) ì†Œìˆ˜ íƒˆë½
            return False
    return True # ëê¹Œì§€ í†µê³¼í•˜ë©´ ì†Œìˆ˜!

print(is_prime(17)) # True
print(is_prime(20)) # False
```
